# Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Line Plot

x = np.arange(10)
y1 = x**2
y2 = 2*x + 3

print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color='green',label="l1",linestyle='dashed')
plt.plot(x,y2, label='l2',marker='*')
plt.legend()
plt.xlabel('X --->')
plt.ylabel('Y --->')
plt.title('My title')
plt.show()
```

```python
#Scatter Plots
```

```python
plt.scatter(x,y1)
plt.show()
```

```python
# Bar Graph
```

```python
plt.bar([1,2,3],[20,30,40],tick_label=['l1','l2','l3'])
plt.show()
```

```python
#Pie chart
```

```python
subjects = 'English', 'Hindi', 'Maths', 'Science'
marks = 80, 90, 100, 100

plt.pie(marks,labels=subjects,explode=[0,0,0,0.1])
plt.show()
```

```python
#histogram
```

```python
plt.hist(marks)
plt.show()
```
